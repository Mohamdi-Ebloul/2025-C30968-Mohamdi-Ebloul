## Tri et indexation

### Comment trier un DataFrame par une ou plusieurs colonnes ?

#### Tri croissant  

```python
df.sort_values(by='Colonne', ascending=True, inplace=True)  
```

#### Tri multi-colonnes

```python
df.sort_values(by=['Col1', 'Col2'], ascending=[True, False])    
```

###  Comment sélectionner des lignes selon des conditions sur les colonnes ?

#### Sélection avec condition

```python
df[(df['Col1'] > 10) & (df['Col2'] == 'Oui')]
```

## Filtrage et conditions

### Comment filtrer un DataFrame selon une ou plusieurs conditions ?

```python
df[df['Colonne'] == 'Valeur']
```

### Comment utiliser la méthode query() pour filtrer les données ?

```python
df.query('Colonne > 10 and AutreColonne == "Oui"')
```

### Comment appliquer une fonction personnalisée ligne par ligne ?

```python
def ma_fonction(ligne):  
    return ligne['Col1'] + ligne['Col2']

df['NouvelleColonne'] = df.apply(ma_fonction, axis=1)
```

## Opérations sur les dates

### Comment convertir une colonne en datetime ?

```python
df['Date'] = pd.to_datetime(df['Date'])
```

### Comment extraire l’année, le mois, le jour ou l’heure d’une date ?

```python
df['Année'] = df['Date'].dt.year  
df['Mois'] = df['Date'].dt.month  
df['Jour'] = df['Date'].dt.day  
df['Heure'] = df['Date'].dt.hour  
```

### Comment filtrer un DataFrame selon une plage de dates ?

```python
df[(df['Date'] >= '2024-01-01') & (df['Date'] <= '2024-12-31')]
```

### Comment créer une colonne avec le nom du jour de la semaine ?

```python
df['JourSemaine'] = df['Date'].dt.day_name()
```

### Comment définir la colonne date comme index du DataFrame ?

```python
df.set_index('Date', inplace=True)
```

## Opérations avancées sur les colonnes

### Comment créer une nouvelle colonne en combinant ou calculant à partir d’autres colonnes ?

```python
df['Total'] = df['Prix'] * df['Quantité']
```

### Comment utiliser np.where pour créer des catégories conditionnelles ?

```python
import numpy as np  
df['Categorie'] = np.where(df['Valeur'] > 100, 'Grand', 'Petit')
```

### Comment mapper des valeurs catégorielles à partir d’un dictionnaire ?

```python
mapping = {'A': 'Excellent', 'B': 'Bon'}  
df['Note'] = df['Code'].map(mapping)
```

### Comment modifier le texte dans une colonne (ex. mettre en minuscules) ?

#### Minuscules

```python
df['Texte'] = df['Texte'].str.lower() 
```

#### Majuscules

```python
df['Texte'] = df['Texte'].str.upper()
```

### Comment séparer une colonne texte en plusieurs colonnes ?

```python
df[['Prénom', 'Nom']] = df['NomComplet'].str.split(' ', expand=True)
```

## Statistiques et fonctions de fenêtres

### Comment calculer la corrélation entre colonnes ?

```python
df.corr()  
df[['Col1', 'Col2']].corr()
```

### Comment calculer des sommes ou produits cumulés ?

```python
df['SommeCumulée'] = df['Valeur'].cumsum()  
df['ProduitCumulé'] = df['Valeur'].cumprod()
```

### Comment calculer des moyennes mobiles (rolling mean) ?

```python
df['MoyenneMobile'] = df['Valeur'].rolling(window=3).mean()
```

### Comment calculer la différence ou le pourcentage de changement entre lignes ?

```python
df['Différence'] = df['Valeur'].diff()  
df['%Changement'] = df['Valeur'].pct_change()
```

### Comment détecter et filtrer les valeurs aberrantes avec l’IQR ?

```python
Q1 = df['Valeur'].quantile(0.25)  
Q3 = df['Valeur'].quantile(0.75)  
IQR = Q3 - Q1  
df_filtré = df[(df['Valeur'] >= Q1 - 1.5*IQR) & (df['Valeur'] <= Q3 + 1.5*IQR)]
```

### Comment calculer la moyenne, médiane, écart-type et variance d’une colonne ?

```python
df['Valeur'].mean()  
df['Valeur'].median()  
df['Valeur'].std()  
df['Valeur'].var()
```
